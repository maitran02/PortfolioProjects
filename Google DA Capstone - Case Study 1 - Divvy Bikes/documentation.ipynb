## **ANALYZE and VISUALIZE**

### **Key Findings**

41% of current users are casual riders, leaving promising opportunity to transition to annual members.



Electric bikes and classic bikes are the most popular with total 97%, docked bikes are only consumed by casual riders. While annual member favor both electric bikes and classic bikes, casual riders are more inclined to electric bikes though the price is slightly higher.



The average riding duration is 20 minutes, in which casual riders ride more with average duration of 30 minutes, while annual members ride less than total average with 13 minutes. 



Saturday has the most number of times the service is being used, the pattern is quite stable during the week.



Opposite patterns exists among user types. Annual members ride much more times than casual riders during workweek, and the figure decreases slightly during weekend. On the contrary, there is a significant difference in the number of times casual riders use bike share during workweek and weekend. The majority of casual riders actively use bike share service during the weekend and the figure is quite small during workweek.



Cyclistic bike share users is the most active at the end of summer and in the beginning of autumn, while there is a small amount of customers riding during spring and winter.



Upon closer analysis of monthly usage patterns within customer segments, it is evident that there is not a substantial difference in the number of rides between annual members and casual riders during the summer season. However, during the remaining seasons, there is a significant disparity, indicating that casual riders utilize the bike-share services less frequently compared to annual members. It would be prudent to launch a targeted campaign aimed at encouraging casual riders to transition into annual members, particularly at the beginning of summer in April when their usage of the bike-share service is higher.

It is worth noting that although there is a declining trend in the number of rides during the less-demanding seasons, annual members continue to contribute significantly to the overall usage. This observation suggests that the annual member segment heavily relies on the bike-share service as their primary mode of transportation for their daily commutes.



Casual riders, despite accounting for a smaller proportion of overall bike-share service usage, exhibit a significantly longer average duration for each ride compared to annual members. On average, their ride durations range from approximately 25 minutes to 35 minutes. One potential explanation for this discrepancy could be the impact of [pricing plans](https://divvybikes.com/pricing). Annual members enjoy the benefit of unlimited 45-minute rides included in their plan, whereas casual riders with the "Day Pass" pricing plan can ride for up to 3 hours at a fixed cost of $1. In fact, 99.3% of casual riders utilize the bike-share service within the 3-hour timeframe, resulting in no additional charges for them.

Therefore, it becomes evident that if users have a need for longer trips each time, opting for annual plans may not be as beneficial since the duration limit is only 45 minutes per ride. The current annual plan proves advantageous primarily for annual members who take short trips and have multiple short trips throughout the month. On the other hand, casual riders tend to be more leisure-oriented users, engaging in fewer trips overall but with significantly longer durations for each ride.

There is noticeable fluctuation in the average riding duration among casual riders throughout the week. They rent bikes most frequently on Saturdays and Mondays, while the average duration of their rides is relatively shorter during the mid-week period. In contrast, the average riding duration among annual members remains relatively stable, fluctuating between 12 and 14 minutes.



A similar trend can be observed when examining the average riding duration by each customer segment on a monthly basis. The average duration for casual riders is considerably larger compared to annual members, approximately double the amount of time rented. This disparity highlights the leisure-oriented nature of casual riders, who tend to engage in longer rides when utilizing the bike-share service.

Moreover, there is a noticeable fluctuation in the duration among casual riders on a monthly basis. This fluctuation further emphasizes the varying riding patterns and preferences within this segment. In contrast, the average duration among annual members remains relatively stable, indicating a more consistent riding behavior.



There is a clear peak in demand for bike rides during the rush hours of 7:00 to 8:00 a.m. in the morning and 4:00 to 6:00 p.m. in the afternoon. This pattern is particularly pronounced among annual members. The underlying reason for this surge in demand can be attributed to the daily commuting needs of individuals for traveling to work or school.

The higher intensity of this demand during rush hours among annual members suggests that a significant portion of them rely on bikes as their main transportation mode for their daily commute. This finding aligns with the understanding that annual members typically utilize the bike-share service for their regular commuting purposes, as opposed to casual riders who primarily use it for leisure activities.

## **Recommendations**

The disparity in price plans is not substantial. However, for casual riders who do not frequently use the service but tend to have longer trips, opting for the "Day Pass" plan would undoubtedly be advantageous. Therefore, I recommend increasing the price for the single plan to incentivize them to transition to the annual plan. Additionally, Cyclistic should consider introducing an additional plan that offers an extended duration for riders who prioritize their riding experience.

It is advisable to initiate a promotional campaign program at the start of summer. This time period witnesses a significant surge in bike-share service usage, making it an opportune moment to attract more annual members and increase engagement with the service.
