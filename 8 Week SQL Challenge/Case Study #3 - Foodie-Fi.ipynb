## **Datasets**

### Table 1: plans

Customers can choose which plans to join Foodie-Fi when they first sign up.

Basic plan customers have limited access and can only stream their videos and is only available monthly at $9.90

Pro plan customers have no watch time limits and are able to download videos for offline viewing. Pro plans start at $19.90 a month or $199 for an annual subscription.

Customers can sign up to an initial 7 day free trial will automatically continue with the pro monthly subscription plan unless they cancel, downgrade to basic or upgrade to an annual pro plan at any point during the trial.

When customers cancel their Foodie-Fi service - they will have a churn plan record with a null price but their plan will continue until the end of the billing period.

### Table 2: subscriptions

Customer subscriptions show the exact date where their specific plan_id starts.

If customers downgrade from a pro plan or cancel their subscription - the higher plan will remain in place until the period is over - the start_date in the subscriptions table will reflect the date that the actual plan changes.

When customers upgrade their account from a basic plan to a pro or annual pro plan - the higher plan will take effect straightaway.

When customers churn - they will keep their access until the end of their current billing period but the start_date will be technically the day they decided to cancel their service.

```python
%%sql --# How many customers has Foodie-Fi ever had?
SELECT COUNT(DISTINCT customer_id) as num_customers
FROM subscriptions
```

```python
%%sql --# What is the monthly distribution of trial plan start_date values for our dataset - use the start of the month as the group by value
SELECT DATE_PART('month', start_date)::INT as month, count(*)
FROM subscriptions
LEFT JOIN plans 
USING (plan_id)
WHERE plan_name = 'trial'
GROUP BY DATE_PART('month', start_date)
ORDER BY month;
```

```python
%%sql --# 3. What plan start_date values occur after the year 2020 for our dataset? Show the breakdown by count of events for each plan_name
SELECT plan_name, COUNT(start_date)AS count_of_events
FROM subscriptions
LEFT JOIN plans 
USING (plan_id) 
WHERE EXTRACT('year' FROM start_date) > 2020 --# after the year 2020
GROUP BY plan_name 
ORDER BY count_of_events DESC
```

```python
%%sql --# 4. What is the customer count and percentage of customers who have churned rounded to 1 decimal place?
SELECT COUNT(DISTINCT customer_id) AS churned_count, 
       ROUND(COUNT(DISTINCT customer_id)::NUMERIC / (SELECT COUNT(DISTINCT customer_id) FROM subscriptions)::NUMERIC * 100, 1) || '%' AS churn_rate
FROM subscriptions 
WHERE plan_id = 4;
```

```python
%%sql --# 5. How many customers have churned straight after their initial free trial - what percentage is this rounded to the nearest whole number?
--# Use self join 
SELECT COUNT(DISTINCT t1.customer_id) AS churned_after_trial_count, 
    ROUND(COUNT(DISTINCT t1.customer_id)::NUMERIC / (SELECT COUNT(DISTINCT customer_id) as total_customers FROM subscriptions)::NUMERIC * 100) || '%' AS churned_after_trial_rate
FROM subscriptions AS t1
LEFT JOIN plans 
USING (plan_id)
INNER JOIN subscriptions AS t2 
ON t1.start_date + INTERVAL '7 days' = t2.start_date AND 
    t1.customer_id = t2.customer_id 
WHERE t1.plan_id = 0 AND t2.plan_id = 4
```

```python
%%sql 
--# LEAD windows partition 
SELECT COUNT(customer_id), 
    ROUND(COUNT(DISTINCT customer_id)::NUMERIC / (SELECT COUNT(DISTINCT customer_id) as total_customers FROM subscriptions)::NUMERIC * 100) || '%' AS churned_after_trial_rate
FROM (
    SELECT *, 
        LEAD(plan_id, 1) OVER (PARTITION BY customer_id ORDER BY start_date) AS next_plan
    FROM subscriptions) as current_next_plan
WHERE plan_id = 0 AND next_plan = 4
```

```python
%%sql --# 6. What is the number and percentage of customer plans after their initial free trial?
SELECT next_plan_after_trial,   
    COUNT(DISTINCT customer_id) AS num_customers,
    ROUND(COUNT(DISTINCT customer_id)::NUMERIC / (SELECT COUNT(DISTINCT customer_id) FROM subscriptions)::NUMERIC * 100, 1) || '%' AS pct_customers
FROM (
    SELECT subscriptions.*, plan_name AS trial_plan, 
        LEAD(plan_name, 1) OVER (PARTITION BY customer_id ORDER BY start_date) AS next_plan_after_trial
    FROM subscriptions
    LEFT JOIN plans 
    USING (plan_id)
    ) AS current_next_plan_subq
WHERE plan_id=0
GROUP BY next_plan_after_trial
ORDER BY COUNT(DISTINCT customer_id)::NUMERIC / (SELECT COUNT(DISTINCT customer_id) FROM subscriptions)::NUMERIC * 100 DESC
```

```python
%%sql --# 7. What is the customer count and percentage breakdown of all 5 plan_name values at 2020-12-31?
--# find existing customers at the end of 2020
SELECT plan_name, COUNT(DISTINCT subscriptions.customer_id) AS count_customers, 
    ROUND(COUNT(DISTINCT subscriptions.customer_id)::NUMERIC / (SELECT COUNT(DISTINCT customer_id) FROM subscriptions)::NUMERIC * 100, 1) AS pct_customers
FROM subscriptions
INNER JOIN (
    SELECT subscriptions.customer_id, MAX(start_date) AS latest_date
    FROM subscriptions
    LEFT JOIN plans
    USING (plan_id)
    WHERE start_date <= '2020-12-31'
    GROUP BY customer_id) AS max_date 
ON subscriptions.customer_id = max_date.customer_id AND
    subscriptions.start_date = max_date.latest_date
LEFT JOIN plans
USING (plan_id)
GROUP BY plan_name
ORDER BY pct_customers DESC
```

```python
%%sql 
SELECT plan_name, COUNT(DISTINCT customer_id) AS count_customers, 
    ROUND(COUNT(DISTINCT customer_id)::NUMERIC / (SELECT COUNT(DISTINCT customer_id) FROM subscriptions)::NUMERIC * 100, 1) AS pct_customers
FROM (
    SELECT *, 
        ROW_NUMBER() OVER(PARTITION BY customer_id ORDER BY start_date DESC) AS row_num
    FROM subscriptions
    WHERE start_date <= '2020-12-31') AS num_subq
LEFT JOIN plans
USING (plan_id)
WHERE row_num = 1
GROUP BY plan_name 
ORDER BY pct_customers DESC
```

```python
%%sql --# 8. How many customers have upgraded to an annual plan in 2020?
--# upgrade: initially not subscribe as pro annual (plan_id = 3) => ROW_NUMBER() <> 1 
WITH num_cte AS (
    SELECT *, 
        ROW_NUMBER() OVER(PARTITION BY customer_id ORDER BY start_date) AS row_num
    FROM subscriptions
    WHERE EXTRACT(YEAR FROM start_date) = 2020
)
SELECT COUNT(DISTINCT customer_id) AS count_upgraded, 
        ROUND(COUNT(DISTINCT customer_id)::NUMERIC / (SELECT COUNT(DISTINCT customer_id) FROM num_cte)::NUMERIC * 100, 1) AS pct_upgraded
FROM num_cte
WHERE plan_id = 3  AND row_num <> 1
```

```python
%%sql --# 9. How many days on average does it take for a customer to an annual plan from the day they join Foodie-Fi?
--# extract a list of customers who DO upgrade to annual plan
--# 1st joined date => row_num = 1, right join with plan_id = 3 
WITH datediff_cte AS (
        SELECT annual_subq.customer_id, plan_id AS first_plan, start_date AS first_date, annual_plan, annual_date, annual_date - start_date AS datediff
    FROM (
        SELECT *, 
            ROW_NUMBER() OVER(PARTITION BY customer_id ORDER BY start_date) AS row_num
        FROM subscriptions) AS num_subq
    RIGHT JOIN ( 
        SELECT customer_id, plan_id AS annual_plan, start_date AS annual_date 
        FROM subscriptions 
        WHERE plan_id = 3 --# extract a list of customers who DO upgrade to annual plan
        ) AS annual_subq
    ON num_subq.customer_id = annual_subq.customer_id 
    WHERE row_num = 1) --# extract the 1st date when customer joined 
    
SELECT AVG(datediff) AS avg_datediff
FROM datediff_cte
```

```python
%%sql --# 10. Can you further breakdown this average value into 30 day periods (i.e. 0-30 days, 31-60 days etc)
WITH datediff_cte AS (
        SELECT annual_subq.customer_id, plan_id AS first_plan, start_date AS first_date, annual_plan, annual_date, annual_date - start_date AS datediff
    FROM (
        SELECT *, 
            ROW_NUMBER() OVER(PARTITION BY customer_id ORDER BY start_date) AS row_num
        FROM subscriptions) AS num_subq
    RIGHT JOIN (
        SELECT customer_id, plan_id AS annual_plan, start_date AS annual_date 
        FROM subscriptions 
        WHERE plan_id = 3
        ) AS annual_subq
    ON num_subq.customer_id = annual_subq.customer_id 
    WHERE row_num = 1)

SELECT TRUNC(datediff / 30 + 1)::INT * 30 - 29 || ' - ' || TRUNC(datediff / 30 + 1)::INT * 30 || ' days' AS bins_category, 
        COUNT(*) AS number_of_customers, 
        ROUND(AVG(datediff)) AS avg_datediff
FROM datediff_cte
GROUP BY TRUNC(datediff / 30 + 1)::INT * 30 - 29 || ' - ' || TRUNC(datediff / 30 + 1)::INT * 30 || ' days'
ORDER BY avg_datediff
```

```python
%%sql --# 11. How many customers downgraded from a pro monthly to a basic monthly plan in 2020?
--# list of customers who do have subscribe BOTH pro_monthly (plan_id = 2) and basic_monthly (plan_id = 1)
--# DOWNGRADE: date when plan_id = 2 < date when plan_id = 1 
WITH downgrade_cte AS(
    SELECT *,
        ROW_NUMBER() OVER(PARTITION BY customer_id ORDER BY start_date) AS row_num 
    FROM (
        SELECT * 
        FROM subscriptions 
        WHERE --# filter for customers who do have subscribe BOTH pro monthly (plan_id = 2) and basic_monthly (plan_id = 1)
                customer_id IN (
                    SELECT customer_id 
                    FROM subscriptions 
                    WHERE plan_id = 2) 
            AND 
                customer_id IN (
                    SELECT customer_id 
                    FROM subscriptions 
                    WHERE plan_id = 1)
    ) AS subq 
    WHERE plan_id = 2 or plan_id = 1 --# filter for only pro monthly and basic monthly 
    ORDER BY customer_id, start_date)

SELECT customer_id
FROM downgrade_cte
LEFT JOIN plans
USING (plan_id)
WHERE customer_id IN (
    SELECT customer_id 
    FROM downgrade_cte
    WHERE plan_id = 2 AND row_num = 1 
)
ORDER BY customer_id
```

```python
%%sql --# 11. How many customers downgraded from a pro monthly to a basic monthly plan in 2020?
--# list of customers who do have subscribe BOTH pro_monthly (plan_id = 2) and basic_monthly (plan_id = 1)
--# DOWNGRADE: date when plan_id = 2 < date when plan_id = 1 
SELECT COUNT(*) AS num_customers_downgraded
FROM (
    SELECT *,
        LEAD(plan_id, 1) OVER (PARTITION BY customer_id ORDER BY start_date) AS next_plan_id
    FROM subscriptions
    LEFT JOIN plans 
    USING (plan_id)
    WHERE start_date <= '2020-12-31'
    ORDER BY customer_id, start_date) AS subq
WHERE plan_id = 2 AND next_plan_id = 1
```

### C. Challenge Payment Question

The Foodie-Fi team wants you to create a new payments table for the year 2020 that includes amounts paid by each customer in the subscriptions table with the following requirements:

monthly payments always occur on the same day of month as the original start_date of any monthly paid plan
upgrades from basic to monthly or pro plans are reduced by the current paid amount in that month and start immediately
upgrades from pro monthly to pro annual are paid at the end of the current billing period and also starts at the end of the month period
once a customer churns they will no longer make payments

```python
%%sql 
DROP TABLE IF EXISTS payment;
CREATE TABLE payment AS
WITH lead_cte AS (
    SELECT customer_id, plan_id, plan_name, start_date, price, 
        LEAD (plan_id, 1) OVER (PARTITION BY customer_id ORDER BY start_date) AS next_plan_id --# find next plan by each customer,
    FROM subscriptions
    LEFT JOIN plans
    USING (plan_id)
    WHERE EXTRACT (year FROM start_date) = 2020 AND plan_id <> 0), --# filter for the year 2020 only and exclude trial plan since no revenue generated 

    self_join_cte AS (
        SELECT t1.customer_id, t1.plan_id, t1.plan_name, t1.start_date, t1.price, t1.next_plan_id,
                t2.plan_name AS next_plan_name, t2.start_date AS next_start_date, t2.price AS next_price
    FROM lead_cte AS t1
    LEFT JOIN lead_cte AS t2 
    ON t1.customer_id = t2.customer_id AND 
        t1.next_plan_id = t2.plan_id
    WHERE t1.plan_id <> 4 --#  
    ORDER BY t1.customer_id), 

    generate_payment_date AS (         
    SELECT *, 
        generate_series(start_date, COALESCE(next_start_date-1, '2020-12-31'), '1 month')::date AS payment_date
    FROM self_join_cte), 

    lag_payment_date AS (
        SELECT *, 
            LAG(payment_date, 1) OVER (PARTITION BY customer_id ORDER BY payment_date) AS previous_payment_date,
            LAG(price, 1) OVER (PARTITION BY customer_id ORDER BY payment_date) AS previous_price
        FROM generate_payment_date),

    row_num_cte AS (
        SELECT customer_id, plan_id, plan_name, payment_date, 
        CASE 
            WHEN payment_date < previous_payment_date + INTERVAL '1 month' THEN price - previous_price
            ELSE price 
        END AS amount, 
        ROW_NUMBER() OVER(PARTITION BY customer_id ORDER BY payment_date) AS payment_order
    FROM lag_payment_date)

SELECT  *
FROM row_num_cte 
WHERE not (plan_id = 3 AND payment_order <> 1) 
ORDER BY customer_id, payment_date;

SELECT *
FROM payment
LIMIT 20;
```

```python
%%sql
DROP TABLE IF EXISTS payments;
CREATE TABLE payments AS
WITH lead_cte AS (
    SELECT customer_id, plan_id, plan_name, start_date, price, 
        LEAD (start_date, 1) OVER (PARTITION BY customer_id ORDER BY start_date) AS next_start_date --# find next start_date by each customer,
    FROM subscriptions
    LEFT JOIN plans
    USING (plan_id)
    WHERE EXTRACT (year FROM start_date) = 2020 AND plan_id <> 0), --# filter for the year 2020 only and exclude trial plan since no revenue generated 

    generate_payment_date AS (         
    SELECT *, 
        generate_series(start_date, COALESCE(next_start_date-1, '2020-12-31'), '1 month')::date AS payment_date --# if no next_start_date => create date until the end of 2020
    FROM lead_cte
    WHERE plan_id <> 4), --# churned subscribers bring no value => remove it
    --# the reason we don't remove churned subscribers simultaneously with trial is that we need the date subscribers stop using the service (LEAD) 

    --# find the previous date when customers paid, if new plan is started immediately (upgrades from basic to monthly or pro plans) and the previous amount paid and upgraded amount paid is within a month => price is reduced
    lag_payment_date AS (
        SELECT *, 
            LAG(payment_date, 1) OVER (PARTITION BY customer_id ORDER BY payment_date) AS previous_payment_date, 
            LAG(price, 1) OVER (PARTITION BY customer_id ORDER BY payment_date) AS previous_price
        FROM generate_payment_date),

    --# create two columns: amount and payment_order
    row_num_cte AS (
        SELECT customer_id, plan_id, plan_name, payment_date, 
        CASE 
            WHEN payment_date < previous_payment_date + INTERVAL '1 month' THEN price - previous_price
            ELSE price 
        END AS amount, 
        ROW_NUMBER() OVER(PARTITION BY customer_id ORDER BY payment_date) AS payment_order
    FROM lag_payment_date)

SELECT  *
FROM row_num_cte 
WHERE not (plan_id = 3 AND payment_order <> 1) --# if customers choose pro annual => they only have to pay once during 2020
ORDER BY customer_id, payment_date;

SELECT *
FROM payments
LIMIT 20;
```
